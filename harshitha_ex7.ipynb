```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
Cars_df=pd.read_csv('mtcars (4).csv')
print(Cars_df)
a=plt.scatter(Cars_df['mpg'],Cars_df['wt'],marker='*')
print(a)
plt.xlabel('mpg')
plt.ylabel('weight')
plt.title('relationship between mpg of the cars and mpg')
plt.show()
trans=Cars_df['am'].value_counts()
print(trans)
trans.plot.bar()
plt.xlabel("transmission")
plt.title("bar plot to check the frequency discribution of transmission type of cars")
plt.show()
Cars_df['mpg'].plot(kind='box')
plt.xlabel('MPG')
plt.title('box and whisker plot of MPG')
plt.show()
min_value=np.min(Cars_df['mpg'])
q1=np.percentile(Cars_df['mpg'],25)
median=np.median(Cars_df['mpg'])
q3=np.percentile(Cars_df['mpg'],75)
max_value=np.max(Cars_df['mpg'])
print('minimum_value:',min_value)
print('Q1:',q1)
print('median:',median)
print('Q3:',q3)
print('maximum_value:',max_value)
```
