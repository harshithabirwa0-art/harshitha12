```python
import pandas as pd
import numpy as np
data={
    'First':['aryan','rohan','riya','yash','siddini'],
    'Lastname':['singh','agarwal','shah','batia','khan'],
    'Type':['full-time employee','intern','full-time employee','part-time employee','full-time employee'],
    'Department':['administration','technical','administration','technical','technical'],
    'Yoe':[2,3,5,7,6],
    'Salary':[20000,5000,10000,20000,4000],
    }
df=pd.DataFrame(data)
print(df)
table=pd.pivot_table(df,index=['Type','Department'],values='Salary',aggfunc=['mean'])
print(table)
tabel1=pd.pivot_table(df,index=['Type','Department'],values='Salary',aggfunc=['sum','mean'])
print(tabel1)
table2=pd.pivot_table(df,index=['Type'],values='Salary',aggfunc=['count'])
print(table2)
table3=pd.pivot_table(df,index=['Type'],values='Salary',aggfunc=['std'])
print(table3)
```
